# E a primeira coisa que temos que fazer é a separação em treino, teste e validação!

 - Buscamos encontrar um **padrão** nos dados que nos ajude a prever a variável resposta
   - Essa é a função do <u>modelo</u>! Ele vai encontrar esses padrões pra gente!
 - Mas para sabermos se um padrão é **útil** precisamos saber se ele é capaz de **generalizar**!
   - É fácil ser "engenheiro de obra pronta": Dar mil explicações pro que aconteceu só **depois** que aconteceu
   - Queremos um modelo que seja capaz de fazer boas predições olhando para o **futuro**, e com dados **diferentes** daqueles que usamos para treinar!
   
   
 - Testar o modelo nos mesmo dados que usamos para treinar (encontrar os padrões) é como "colar na prova": 
   - Você prova que você decorou o livro
   - Mas não aprende a usar aqueles conceitos em situações do mundo real

```python
df_dev = df.loc[(df.anomesdia<'2017-01-01')].set_index(['SiteId','anomesdia'])
df_val = df.loc[(df.anomesdia>='2017-01-01')].set_index(['SiteId','anomesdia'])

X_dev = df_dev.drop(['Value','Value_bool'],axis=1)
y_dev = df_dev.Value_bool.astype(int)

X_val = df_val.drop(['Value','Value_bool'],axis=1)
y_val = df_val.Value_bool.astype(int)

from sklearn.model_selection import train_test_split
X_tr, X_ts, y_tr, y_ts = train_test_split(X_dev, y_dev, test_size=.5, random_state=61658)

X_tr
```

```python
print( y_tr.mean() )
print( y_ts.mean() )
print( y_val.mean() )
```

 - No treino/teste é <u>esperado</u> que os valores dêem próximos:
   - Foi uma amostragem aleatória do mesmo conjunto
   
   
 - Já na validação, **idealmente os valores <u>deveriam</u> ser próximos**.
   - Mas nós não controlamos o mundo real!
   - Variar quer dizer que nossa variável resposta **não está estacionária** (ela muda com o passar do tempo)
   - Isso indica que nosso modelo provavelmente se **descalibrará depois de um tempo**, exigindo manutenção
     - Nota: Todo modelo descalibra. O mundo muda. Nós precisamos **monitorar e acompanhar**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

params = {
    'max_leaf_nodes': 2**np.arange(2,12),
    'criterion': ['gini','entropy'],
    'class_weight': [None, 'balanced']
}

grid = GridSearchCV(
    DecisionTreeClassifier(random_state=61658), # Primeiro argumento: O modelo que vamos usar
    params, # Segundo argumento: a lista de parâmetros que vamos testar
    cv=10, # Argumento `cv`: Em quantas partes vamos "fatiar" nosso dataset
    scoring='roc_auc', # Argumento `scoring`: Qual métrica estamos maximizando?
    verbose=10, # Aqui é só pra ele mostrar na tela
    n_jobs=4, # Se você tiver mais processadores, você pode aumentar até o número que você tiver
)

#### CUIDADO!!!!
# Não use mais n_jobs do que a quantidade de processadores que você tem!
# Nota: Se você usar mais de 1 job, ele não mostra a atualização aqui no notebook
# Mas ele vai mostrar no seu terminal que você abriu o Jupyter

grid.fit(X_tr,y_tr)

from sklearn.metrics import roc_auc_score
print( roc_auc_score(y_ts , grid.predict_proba(X_ts )[:,1]) )
print( roc_auc_score(y_val, grid.predict_proba(X_val)[:,1]) )
```

# Mas será que um modelo tão "simples" é um bom benchmark?

 - Upgrade "simples": **Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

params = {
    'max_leaf_nodes': 2**np.arange(2,12),
    'criterion': ['gini','entropy'],
    'class_weight': [None, 'balanced']
}

grid2 = GridSearchCV(
    RandomForestClassifier(n_estimators=700, random_state=61658), # Primeiro argumento: O modelo que vamos usar
    params, # Segundo argumento: a lista de parâmetros que vamos testar
    cv=10, # Argumento `cv`: Em quantas partes vamos "fatiar" nosso dataset
    scoring='roc_auc', # Argumento `scoring`: Qual métrica estamos maximizando?
    verbose=10, # Aqui é só pra ele mostrar na tela
    n_jobs=8, # Se você tiver mais processadores, você pode aumentar até o número que você tiver
)

#### CUIDADO!!!!
# Não use mais n_jobs do que a quantidade de processadores que você tem!
# Nota: Se você usar mais de 1 job, ele não mostra a atualização aqui no notebook
# Mas ele vai mostrar no seu terminal que você abriu o Jupyter

grid2.fit(X_tr,y_tr)

from sklearn.metrics import roc_auc_score
print( roc_auc_score(y_ts , grid2.predict_proba(X_ts )[:,1]) )
print( roc_auc_score(y_val, grid2.predict_proba(X_val)[:,1]) )
```

```python
print( grid2.best_params_ )
```

```python
X_ts_pred2 = X_ts.copy()
X_ts_pred2.loc[:,'pred'] = grid2.predict_proba(X_ts)[:,1]
X_ts_pred2.loc[:,'target'] = y_ts
X_ts_pred2 = X_ts_pred2.reset_index()
X_ts_pred2.loc[:,'anomes'] = pd.to_datetime( X_ts_pred2.anomesdia.dt.strftime('%Y-%m-01') )

X_val_pred2 = X_val.copy()
X_val_pred2.loc[:,'pred'] = grid2.predict_proba(X_val)[:,1]
X_val_pred2.loc[:,'target'] = y_val
X_val_pred2 = X_val_pred2.reset_index()
X_val_pred2.loc[:,'anomes'] = pd.to_datetime( X_val_pred2.anomesdia.dt.strftime('%Y-%m-01') )
```

```python
anomeses_ts2 = np.sort( X_ts_pred2.anomes.unique() )
aucs_ts2 = []
target_ts2 = []
for anomes in anomeses_ts2:
    print(anomes)
    X_loc = X_ts_pred2.loc[X_ts_pred2.anomes==anomes]
    
    target_ts2.append( X_loc.target.mean() )
    aucs_ts2.append( roc_auc_score(X_loc.target, X_loc.pred) )

    
    
anomeses_val2 = np.sort( X_val_pred2.anomes.unique() )
aucs_val2 = []
target_val2 = []
for anomes in anomeses_val2:
    print(anomes)
    X_loc = X_val_pred2.loc[X_val_pred2.anomes==anomes]
    
    target_val2.append( X_loc.target.mean() )
    aucs_val2.append( roc_auc_score(X_loc.target, X_loc.pred) )
```

```python
plt.figure(figsize=(15,5))

plt.plot(anomeses_ts2, aucs_ts2, 'o-')
plt.plot(anomeses_val2, aucs_val2, 'o-')

plt.grid();

plt.figure(figsize=(15,5))

plt.plot(anomeses_ts2, target_ts2, 'o-')
plt.plot(anomeses_val2, target_val2, 'o-')

plt.grid();
```

# Para qualquer limiar que escolhermos do nosso modelo, temos prós e contras:

 - Diremos que, acima de um certo limiar, aquele local irá consumir mais do que a média
   - Dos acima da média que acertamos, chamamos de **verdadeiros positivos**
   - Entretanto, há vários locais que consumiram pouco, mesmo com limiares altos. Esses são os **falsos positivos**
   
   
   - Além disso, entre os que ficaram abaixo do limiar, há muitos locais que realmente consumiram dentro do esperado. Esses são os **verdadeiros negativos**
   - Entretanto, há locais de alto consumo também, os **falsos negativos**
   
   
### Vamos concentrar nossa atenção nos verdadeiros positivos e nos falsos positivos (lado acima do limiar)

 - Se colocamos o limiar mais pra baixo, aumentamos a **taxa de verdadeiros positivos**, o que é bom!
 - Mas também aumentamos a **taxa de falsos positivos**! O que é ruim!
 
 
 - Sempre para **capturar amostras positivas** <u>pagamos um preço</u> de **classificar erroneamente algumas amostras negativas**!
 
 
### Como medimos esse preço???

```python
from sklearn.metrics import roc_curve

fpr, tpr, ths = roc_curve(X_val_pred2.target, X_val_pred2.pred)

plt.figure(figsize=(10,8))
plt.plot(fpr, tpr, 'k-', lw=1., zorder=1)
plt.scatter(fpr, tpr, c=ths, s=30., vmin=0., vmax=1., zorder=20, cmap='tab20c')
plt.colorbar()

plt.plot((0,1), (0,1), color='k', lw=5.)

plt.xlabel('Taxa de falsos positivos', fontsize=16)
plt.xticks(fontsize=16)
plt.ylabel('Taxa de verdadeiros positivos', fontsize=16)
plt.yticks(fontsize=16)

plt.grid();
```
